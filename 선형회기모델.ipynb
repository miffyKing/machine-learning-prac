```python
height = np.array([183,150, 180, 197, 160, 175])
print(height)
```

```python
height = height.reshape(-1,1)   #2차원값 들어가야해서
print(height)
```

```python
math = np.array([85,45,80,99,45, 75])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
line_fitter = LinearRegression()
print(line_fitter)
```

```python
line_fitter.fit(height, math)           #키와 수학성적간의 상관관계를 linear regression 으로 구해볼까 해

#fit()함수 : line_fitter.coef_ : 기울기저장
#           line_fitter.intercept_ : 절편 저장
```

```python
score_predict = line_fitter.predict(height)
```

```python
plt.plot(height, math, 'x')
```

```python
plt.plot(height, score_predict)
plt.show()
```

```python
line_fitter.coef_     #기울기
```

```python
line_fitter.intercept_      #절편
```

```python
#성능 평가
from sklearn.metrics import mean_squared_error        #성능평가를 위해 mse 를 사용하겠다!
```

```python
print("MEan Squared Error  : ", mean_squared_error(score_predict, math))
```

```python
print(" RMSE MEan Squared Error  : ", mean_squared_error(score_predict, math)**0.5 )
```

```python
print('score : ' , line_fitter.score(height, math))
```
