```python
X = iris.data
y = iris.target 
# X가 iris data, y가 target data, 꽃 종ㅇ류
```

```python
df = pd.DataFrame(X, columns = ['sepal_witdh(cm)', 'sepal_length(cm)', 'petal_width(cm)', 'petal_length(cm)'])    
#4개를 데이터 프레임으로 묶음
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['species'] = np.array([iris.target_names[i] for i in iris.target])

sns.pairplot(df, hue='species')
#seaborn 으로 그래프 시각화 , hue = 'species' -> 이 species 를 기준을 ㅗ분류하겠다
#3차원 이상 데이터 에서 주로 사용
#hue = 'species'->카테고리형 데이터가 섞여 있을 때 이걸로 카테고리 값에 따라 색상을 달리함
```

```python
 from sklearn.model_selection import train_test_split     #data나누는 거
 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
#70프로는 training data
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_r = LogisticRegression()
```

```python
log_r.fit(X,y)
```

```python
W, b = log_r.coef_, log_r.intercept_
```

```python
W ,b
```

```python
print(log_r.score(X_train, y_train))
```

```python
#X,y 둘다 train 데이터로 훈련시켰을 때의 값
```
