```python
# dictionary 형태를 데이터 형으로.. 이 dic의 key 값이 바로 column 명!
dic1 = {'Name' : ['Byounghwa', 'Seohwi', 'Joonbaem', 'Hoonbalnom'],
        'Year' : ['21', '20', '19','5'],
        'Gender' : ['M','M', 'M','W']
        }
data_df = pd.DataFrame(dic1)

print(data_df)
#그냥 알아서 DataFrame으로 바꿔주고, 이를 통해 shape, index 같은 정보를 쉽게 얻어낼 수가 있노
print(data_df.index)
print('ㅗ'*30)

print(data_df.shape)
print('#'*30)

data_df = pd.DataFrame(dic1, columns = ['Names', 'Year', 'Gender' , 'Age'])
print(data_df)
print('#'*30)
data_df = pd.DataFrame(dic1, index = ['one', 'two', 'three', 'four'])
print(data_df)
#data_df = pd.DataFrame(dic1,  = ['11', '12', '13', '14']) NaN으로 되어있는 Age에 값을 채우려면 어떻게 해야하노? .. 모르겠다...
print('#'*30)
print(data_df)
```

```python
#value_counts - 특절 column의 데이터 값 분포를 확인하고 싶을때 주로 사용. 동일한 개별 데이터 값이 몇건이 있는지에 대한 정보 제공, 
#주의할 점은 value counts() 는 Series 객체에서만 호출 될 수 있기 때문에 반드시 DataFrame을 단일컬럼으로 입력하여 Series로 변환 후 호출해야 함.
```

```python
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
```

```python
series1 = titanic_df['Name']
value_count2 = series1.value_counts()
#이렇게 시리즈로 데이터프레임을 바꿔놓고 사용해야만 해요..
```

```python
print(value_count2)
```

```python
#Sort_Values() - 정렬ㄹ컬럼. ascending = True, False 로 오름,내림 차순 순으로 정렬할 수 있어요
```
